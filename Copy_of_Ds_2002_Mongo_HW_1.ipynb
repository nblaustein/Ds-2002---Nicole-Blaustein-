```python
#3.1
avg_query = [{"$group": {"_id": None, "averageAwards": {"$avg": "$awards.wins"}}}]
average_awards = list(collection.aggregate(avg_query))

# Print the average number of awards
for document in average_awards:
    print(f"Average Awards: {document['averageAwards']}")

```

```python
#3.2
pipeline = [
    {
        "$group": {
            "_id": "$title",
            "maxAwards": {"$max": "$awards.wins"}
        }
    },
    {
        "$sort": {"maxAwards": -1}
    },
    {
        "$limit": 1
    },
    {
        "$project": {"_id": 0, "title": "$_id", "maxAwards": 1}
    }
]

# Execute the aggregation pipeline
result = list(collection.aggregate(pipeline))

df = pd.DataFrame(result)
print(df)

```
